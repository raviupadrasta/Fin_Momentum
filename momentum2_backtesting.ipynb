```python
#import libraries
import yfinance as yf
import numpy as np
import pandas as pd
from pandas.tseries.offsets import MonthEnd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stocks = pd.read_html('https://en.wikipedia.org/wiki/NIFTY_500')[2]
stocks_data = stocks[3]
stocks_data = stocks_data + '.NS'
stocks_data = stocks_data.to_list()
stocks_data
```

```python
#pulling price data
start = '2024-01-01'

df = yf.download(stocks_data, start = start)['Close']

"""prices, symbols = [],[]

for symbol in stocks_data:
    df_ = yf.download(symbol, start = start)['Close']
    if not df_.empty:
        prices.append(df)
        symbols.append(symbol)"""
```

```python
#calculate daily percentage change
df.pct_change()

#cummulate the daily returns
(df.pct_change() + 1).prod() -1

#cummulate the returns over monthly period
all_monthly_ret = df.pct_change().resample('M').agg(lambda x : (x+1).prod() -1)
```

```python
#aggregating returns on 3 month rolling period
all_monthly_ret_3 = all_monthly_ret.rolling(3).agg(lambda x: (x+1).prod() -1)
```

```python
all_monthly_ret_3.dropna(inplace=True)
all_monthly_ret_3
```
